```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os

data = pd.read_csv("C:/Users/kanch/OneDrive/Documents/MATH331/CNN/exo_data.csv")

labels = data['LABEL'].values
flux_data = data.drop(columns=['LABEL']).values

output_folder = "C:/Users/kanch/OneDrive/Documents/MATH331/CNN/cnn_stars"
os.makedirs(output_folder, exist_ok=True)

existing_images = [f for f in os.listdir(output_folder) if f.endswith('.png')]
existing_indices = {int(f.split('_')[1]) for f in existing_images}

total_rows = flux_data.shape[0]

for i in range(total_rows):
    if i in existing_indices:
        continue 

    star_flux = flux_data[i]
    time = np.arange(len(star_flux))

    plt.figure(figsize=(10,4))
    plt.plot(time, star_flux, linewidth=1.2)
    plt.xlabel("Time")
    plt.ylabel("Flux")
    plt.title(f"Light Curve: {'Exoplanet' if labels[i]==1 else 'Non-Exoplanet'}")
    plt.grid(True)

    output_path = os.path.join(output_folder, f"star_{i}_label_{labels[i]}.png")
    plt.savefig(output_path, dpi=300)
    plt.close()

    print(f"Saved plot for star {i} at {output_path}")
```
